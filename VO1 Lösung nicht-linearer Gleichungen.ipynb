# Nullstellensuche

```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

## Testfunktionen

```python
def x_squared(x):
    return x**2

def x_cube(x, diff=0):
    if(diff == 2):
        return 6*x+2
    elif(diff == 1):
        return 3*x**2+2*x-15
    else:
        return x**3+x**2-15*x
```

## Hilfsfunktionen

```python
def plot_interval(y, x=[], start_lim = 1000, interval = 100):
    end_lim = start_lim + interval
    if(len(x) == 0):
        plt.plot(np.arange(len(y[start_lim:end_lim])), y[start_lim:end_lim])
    else:
        plt.plot(x[start_lim:end_lim], y[start_lim:end_lim])
    plt.show()
```

## Bracketing

```python
x = np.arange(1000)/100-5.01234
y = x_cube(x)

# 0-th derivative
plot_interval(y, x, 0, len(x))
# 1-st derivative
plot_interval(x_cube(x, 1), x, 0, len(x))
# 2-nd derivative
plot_interval(x_cube(x, 2), x, 0, len(x))
```

```python
# select every sel_interval-th value
sel_interval = 50
selection_y = y[::sel_interval]
selection_x = x[::sel_interval]
len(y), len(selection_y), len(selection_x)
```

```python
negative_x = []
for i in range(1, len(selection_y)):
#     print(selection_y[i-1]*selection_y[i])
    if(selection_y[i-1]*selection_y[i] < 0):
        negative_x.append([selection_x[i-1], selection_x[i]])
print("len before bracketing: %i, len after bracketing: %i" % (len(selection_x), len(negative_x)))
print("intervals:\n%s" % str(np.round(np.array(negative_x), 2)))
```

## Bisection

use intervals as starting point for recursive search

```python
for i in range(len(negative_x)):
    curr_interval = negative_x[i]
#     print("difference between upper and lower bracket: %f" % np.diff(curr_interval))

    itr=1
    while(np.diff(curr_interval) > 0.000001):
        itr += 1
#         print("difference between upper and lower bracket: %f" % np.diff(curr_interval))
        mean = np.mean(curr_interval)

        if(x_cube(curr_interval[0])*x_cube(mean) < 0):
            curr_interval[1] = mean
        elif(x_cube(curr_interval[1])*x_cube(mean) < 0):
            curr_interval[0] = mean
        else:
            raise SystemError()

    print("final value: %f in %i iterations" % (np.mean(curr_interval), itr))
```

## Newton-Raphson-Methode

```python
negative_x = []
for i in range(1, len(selection_y)):
    if(selection_y[i-1]*selection_y[i] < 0):
        negative_x.append([selection_x[i-1], selection_x[i]])
print(negative_x)
        
for i in range(len(negative_x)):
    curr_value = negative_x[i][0]
                            # Funktion          # erste Ableitung der Funktion
    next_value = curr_value-(x_cube(curr_value)/x_cube(curr_value, 1))

    itr=1
    while(abs(curr_value-next_value) > 0.000001):
        itr += 1
#         print(abs(curr_value-next_value))
        curr_value = next_value
        next_value = curr_value-(x_cube(curr_value)/x_cube(curr_value, 1))
    
    print("final value: %f in %i iterations" % (next_value, itr))
```
